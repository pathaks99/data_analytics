```python
df['SleepDay'] = pd.to_datetime(df['SleepDay'])
df.head(5)
```

```python
df['NotSleepTime']=df['TotalTimeInBed']-df['TotalMinutesAsleep']
```

```python
df["SleepDay"] = pd.to_datetime(df["SleepDay"])
df.head(5)
```

```python
df.info()
```

```python
df['Year'] = pd.DatetimeIndex(df['SleepDay']).year
df['Month'] = pd.DatetimeIndex(df['SleepDay']).month_name()
df['Weekday'] = pd.DatetimeIndex(df['SleepDay']).day_name()
```

```python
df.Month.unique()
```

```python
plt.figure(figsize = (25,15))
sns.lineplot(x="SleepDay", y="TotalMinutesAsleep",data=df,label="TotalMinutesAsleep")
sns.lineplot(x="SleepDay", y="TotalTimeInBed",data=df,label="TotalTimeInBed")
plt.xlabel("SleepDay", fontsize=18)
plt.ylabel("TotalTimes", fontsize=18)
plt.title("Sleep Day V/s Time ", fontsize = 25)
plt.xticks(rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
plt.figure(figsize = (25,15))
sns.lineplot(x="TotalTimeInBed", y="TotalMinutesAsleep",data=df)
plt.show()
```

```python
plt.figure(figsize = (25,15))
sns.lineplot(x="NotSleepTime", y="TotalSleepRecords",data=df)
plt.show()
```

```python
plt.figure(figsize = (25,15))
sns.barplot(x="Weekday", y="NotSleepTime",data=df)
plt.show()
```

```python
filter_1= df.where(df['Id']==1503960366)
filter_1.dropna(inplace=True)
filter_1.head(5)
```

```python
plt.figure(figsize = (25,15))
sns.lineplot(x="SleepDay", y="TotalTimeInBed", hue="Month",dashes=False, markers=True,data=filter_1)
plt.xlabel("SleepDay", fontsize=18)
plt.ylabel("TotalTimeInBed", fontsize=18)
plt.title("Sleep Day V/s Time in BED for ID=1503960366", fontsize = 25)
plt.xticks(rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.legend(title="Month",fontsize = '20', title_fontsize = "20")


plt.show()
```
